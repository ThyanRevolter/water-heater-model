```python
from pype_schema.parse_json import JSONParser
network = JSONParser("output/dhw_draw_network.json").initialize_network()
```

```python
dwh = network.get_node("DHWSystem")

for node in dwh.get_all_nodes():
    print(node.id, type(node))
for connection in dwh.get_all_connections():
    print(connection.id, type(connection), connection.source, connection.destination)
```
